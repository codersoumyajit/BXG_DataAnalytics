```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv(r"C:\Users\soumy\OneDrive\AppData\Desktop\clean_data_after_eda.csv")
```

```python
X = data.drop(columns=['churn'])
y = data['churn']
```

```python
X_encoded = pd.get_dummies(X, drop_first=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
```

```python
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)
```

```python
importances = pd.Series(model.feature_importances_, index=X_encoded.columns)
```

```python
weakest_feature = importances.idxmin()
print(f"Weakest feature removed: {weakest_feature}")
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv(r"C:\Users\soumy\OneDrive\AppData\Desktop\clean_data_after_eda.csv")

if 'id' in df.columns:
    df.drop(columns=['id'], inplace=True)

low_variance_cols = [col for col in df.columns if df[col].nunique() <= 1]
df.drop(columns=low_variance_cols, inplace=True)

numeric_df = df.select_dtypes(include=['number'])  # this avoids string conversion errors
corr_matrix = numeric_df.corr().abs()

upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

to_drop = [column for column in upper.columns if any(upper[column] > 0.95)]
df.drop(columns=to_drop, inplace=True)

df.head()
```

```python
df.columns.tolist()
```

```python
# Convert all date columns to datetime format
date_columns = ['date_activ', 'date_end', 'date_modif_prod', 'date_renewal']  # update if needed

for col in date_columns:
    df[col] = pd.to_datetime(df[col], errors='coerce')  # safely convert to datetime
    df[f'{col}_year'] = df[col].dt.year
    df[f'{col}_month'] = df[col].dt.month
    df[f'{col}_day'] = df[col].dt.day
    df[f'{col}_dayofweek'] = df[col].dt.dayofweek
    df[f'{col}_dayofyear'] = df[col].dt.dayofyear

# Optionally, drop the original date columns
df.drop(columns=date_columns, inplace=True)
```

```python
df.columns.tolist()
```

```python
df.to_csv('cleaned_dataset.csv', index=False)
```
